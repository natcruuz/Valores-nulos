Aquí verificamos qué columnas de información tenían datos nulos y cuántos tenían.

**Primer método de sustitución de valores nulos**

Realizamos una copia de la tabla de datos inicial para posteriormente llenar los valores nulos de la columna "salon_ventas" con el promedio de los valores de esa misma columna.

```python
#Copiamos los valores de la tabla inicial
data1 = data.copy()
data1
```

```python
#Primer método de sustitución de valores nulos
#Sustituir valores nulos con promedio o media
data1["salon_ventas"] = data1["salon_ventas"].fillna(round(data["salon_ventas"].mean(),1))
data1
```

```python
#Corroporamos valores nulos
valores_nulos = data1.isnull().sum()
valores_nulos
```

**Segundo método de sustición de variables**

Para este método sustituimos los valores nulos de la columna "tarjeta_debito" con la mediana de los datos de esta misma columna.

```python
#Segundo método de sustitución de valores nulos
#Sustituir valores nulos con mediana
data1["tarjetas_debito"] = data1["tarjetas_debito"].fillna(round(data["tarjetas_debito"].median(),1))
```

```python
#Corroboramos valores nulos
valores_nulos = data1.isnull().sum()
valores_nulos
```

**Tercer método de sustición de variables**

En este método sustituimos los valores nulos de la columna "tarjetas_credito" por un valor en específico, que en este caso fue 945886671.2

```python
#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en concreto
data1["tarjetas_credito"] = data1["tarjetas_credito"].fillna(945886671.2)
```

```python
#Corroboramos valores nulos
valores_nulos = data1.isnull().sum()
valores_nulos
```

**Cuatro método de sustitución de valores nulos** (aquí ya se pueden variables de tipo categórico)

Este método sustituye valores nulos por algún string en concreto, por lo que no fue empleado en este código de datos numéricos.

**Quinto método de sustición de variables**

Con este método se sustituyó cada uno de los valores nulos de la columna "otros medios" por el valor de la fila de arriba.

```python
#Sustituir los valores nulos por valores no nulos hacia adelante "forward fill" ("ffill")
#Filtro por columnas
data1["otros_medios"] = data1["otros_medios"].fillna(method="ffill")
```

```python
#Corroboramos valores nulos
valores_nulos = data1.isnull().sum()
valores_nulos
```

**Sexto método de sustitución de valores nulos**

Con este método se sustituye cada uno de los valores nulos de la columna "subtotal_ventas_alimentos_bebidas" con el valor de la fila inferior.

```python
#Sustituir valores nulos por valores no nulos hacia atrás backwardfill ("bfill")
#Filtro por columnas
data1["subtotal_ventas_alimentos_bebidas"]=data1["subtotal_ventas_alimentos_bebidas"].fillna(method="bfill")
```

```python
#Corroboramos los valores nulos
valores_nulos = data1.isnull().sum()
valores_nulos
```
